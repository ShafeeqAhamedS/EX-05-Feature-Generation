# Apply feature scaling for numerical data

```python
df2 = df.copy()
```

```python
df2 = df.drop(['car_ID', 'CarName', 'fueltype', 'aspiration', 'doornumber', 'carbody','drivewheel', 'enginelocation', 'enginetype', 'cylindernumber','fuelsystem'],axis=1)
```

```python
df2
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
df_minmax=pd.DataFrame(sc.fit_transform(df2),columns=['wheelbase','carlength','carwidth','carheight','curbweight','enginesize'])
df_minmax
```

```python
from sklearn.preprocessing import StandardScaler
sc1=StandardScaler()
df_std=pd.DataFrame(sc1.fit_transform(df2),columns=['wheelbase','carlength','carwidth','carheight','curbweight','enginesize'])
df_std
```

```python
from sklearn.preprocessing import MaxAbsScaler
sc2=MaxAbsScaler()
df_max=pd.DataFrame(sc2.fit_transform(df2),columns=['wheelbase','carlength','carwidth','carheight','curbweight','enginesize'])
df_max
```

```python
from sklearn.preprocessing import RobustScaler
sc3=RobustScaler()
df_rob=pd.DataFrame(sc3.fit_transform(df2),columns=['wheelbase','carlength','carwidth','carheight','curbweight','enginesize'])
df_rob
```
